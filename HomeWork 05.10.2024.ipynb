# Some Data Virtualizaion

### Visualize Missing Data

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize missing data
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Visualization')
plt.show()
```

```python
# Count of survival
sns.countplot(x='survived', data=df)
plt.title('Survival Count')
plt.show()

# Survival rate in percentage
df['survived'].value_counts(normalize=True)
```

```python
# Distribution of survival based on gender
sns.countplot(x='survived', hue='sex', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
# Age distribution
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

# Age and survival
sns.histplot(df[df['survived'] == 1]['age'].dropna(), bins=30, kde=True, color='green', label='Survived')
sns.histplot(df[df['survived'] == 0]['age'].dropna(), bins=30, kde=True, color='red', label='Did Not Survive')
plt.legend()
plt.title('Age vs Survival')
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=df)
plt.title('Survival by Class')
plt.show()

# Survival rates
df.groupby('pclass')['survived'].mean()
```

```python
sns.countplot(x='parch', hue='survived', data=df)
plt.title('Survival by Class')
plt.show()

# Survival rates
df.groupby('parch')['survived'].mean()
```
